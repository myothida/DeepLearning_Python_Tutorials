```python
import pandas as pd

text = """On Thursday, Astronomer CEO Byron was shown romantically
 holding his company's chief people officer,  \
Kristin Cabot, at a recent Coldplay concert."""


from utils import text_helper
from utils.tokenizer import CharTokenizer, WordTokenizer, NGramsTokenizer
word_tokenizer = WordTokenizer.train_from_text(text)
encoded = word_tokenizer.encode(text)
encoded_df = pd.DataFrame(encoded, columns=['word'])
tokens = word_tokenizer.word_for_token_id
tokens =  list(tokens.values())
print("Tokens:",tokens)
```

```python
import nltk
#nltk.download('averaged_perceptron_tagger_eng', download_dir='./.venv/nltk_data')
#nltk.download('tagsets_json', download_dir='./.venv/nltk_data')
nltk.data.path.clear()  
nltk.data.path.append("D:/09_Projects/04_DeepLearning/DeepLearning_Python_Tutorials/.venv/nltk_data")
print(nltk.data.path)


from nltk.data import find
try:
    find('taggers/averaged_perceptron_tagger_eng')
    print("Tagger model found ✅")
except LookupError as e:
    print("❌ Model not found:", e)
    raise

try:
    pos_tags = nltk.pos_tag(tokens)
    print("POS tags:", pos_tags)
except Exception as e:
    print("❌ ERROR DURING pos_tag():")
    print(type(e).__name__ + ":", e)

pos_descriptions = {
    'NN': 'noun (singular or mass)',
    'NNS': 'noun (plural)',
    'NNP': 'proper noun (singular)',
    'NNPS': 'proper noun (plural)',
    'PRP': 'personal pronoun',
    'PRP$': 'possessive pronoun',
    'VB': 'verb (base form)',
    'VBD': 'verb (past tense)',
    'VBG': 'verb (present participle or gerund)',
    'VBN': 'verb (past participle)',
    'VBP': 'verb (non-3rd person singular present)',
    'VBZ': 'verb (3rd person singular present)',
    'JJ': 'adjective',
    'JJR': 'adjective (comparative)',
    'JJS': 'adjective (superlative)',
    'RB': 'adverb',
    'RBR': 'adverb (comparative)',
    'RBS': 'adverb (superlative)',
    'DT': 'determiner',
    'WDT': 'wh-determiner',
    'WP': 'wh-pronoun',
    'WP$': 'possessive wh-pronoun',
    'WRB': 'wh-adverb',
    'IN': 'preposition or subordinating conjunction',
    'CC': 'coordinating conjunction',
    'EX': 'existential there',
    'MD': 'modal verb',
    'POS': 'possessive ending',
    'TO': 'to (as in "to go")',
    'UH': 'interjection',
    'CD': 'cardinal digit',
    'LS': 'list item marker',
    'FW': 'foreign word',
    'RP': 'particle',
    'SYM': 'symbol',
    'PDT': 'predeterminer',
    '.': 'punctuation (sentence ending)',
    ',': 'punctuation (comma)',
    ':': 'punctuation (colon or semi-colon)',
    '(': 'punctuation (opening parenthesis)',
    ')': 'punctuation (closing parenthesis)',
    'HYPH': 'hyphen',
    'LS': 'list item marker',
}
```

```python
df_pos = pd.DataFrame(pos_tags, columns=['word', 'pos'])
df_pos['description'] = df_pos['pos'].map(pos_descriptions)
df_pos['is_noun'] = df_pos['pos'].str.startswith('NN')
```

```python
entity_definitions = {
    "PER": "Person - Named individuals like people or fictional characters",
    "ORG": "Organization - Companies, institutions, teams, etc.",
    "LOC": "Location - Geographical places (mountains, rivers, etc.)",
    "GPE": "Geo-Political Entity - Countries, cities, states",
    "MISC": "Miscellaneous - Events, nationalities, works of art, etc.",
    "DATE": "Date or temporal expression",
    "TIME": "Time expression",
    "FAC": "Facility - Buildings, airports, highways",
    "PRODUCT": "Product - Physical or digital product names",
    "EVENT": "Event - Named hurricanes, battles, etc.",
    "LANGUAGE": "Language - English, Spanish, etc.",
}
```

```python
from transformers import pipeline

ner = pipeline("ner",model="dbmdz/bert-large-cased-finetuned-conll03-english", revision="4c53496",aggregation_strategy="simple")

results = ner(text)
results_df = pd.DataFrame(results)

results_df['entity_description'] = results_df['entity_group'].map(entity_definitions)
results_df

```
